# yfinance Playground

This file uses the yfinance library to pull stock data for a specific stock

```python
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
```

```python
clsk = yf.Ticker("CLSK")
```

```python
end_date = datetime.now().strftime('%Y-%m-%d')
day_history = clsk.history(start='2022-01-01', end=end_date, interval="1mo")
```

```python
day_history
```

```python
fig = go.Figure(data=go.Scatter(x=day_history.index,y=day_history['Close'], mode='lines'))
fig.show()
```

```python
from plotly.subplots import make_subplots

fig3 = make_subplots(specs=[[{"secondary_y": True}]])
fig3.add_trace(go.Candlestick(x=day_history.index,
                              open=day_history['Open'],
                              high=day_history['High'],
                              low=day_history['Low'],
                              close=day_history['Close'],
                             ))
```
